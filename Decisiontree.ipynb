```python
#Building and Visualizing a Decision Tree

#importing Necessary Libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

#loading the dataset
wine=load_wine()
X=wine.data #Features
y=wine.target #target variable

#splitting the data into training and test sets
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)

#creating and fitting the model
dt_clf=DecisionTreeClassifier(criterion='gini',max_depth=3,random_state=42)
dt_clf.fit(X_train,y_train)

#making predictions
y_pred=dt_clf.predict(X_test)

#evaluating the model
accuracy=accuracy_score(y_test,y_pred)
classification=classification_report(y_test,y_pred)
confusion_mat=confusion_matrix(y_test,y_pred)

#Displaying the results
print(f"Accuracy Score:{accuracy}")
print("Classification Report:")
print(classification)
print("Confusion Matrix:")
print(confusion_mat)

#visualizing the tree
plt.figure(figsize=(15,10))
plot_tree(dt_clf,feature_names=wine.feature_names,class_names=wine.target_names,filled=True)
plt.show()
```
